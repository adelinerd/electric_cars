```python
relevant_columns = [ 'Erstzulassung', 'Postleitzahl', 'Standort',
       'Geburtsjahr Halter', 'Kraftstoffart', 'Leistung (kW)']
df = df [relevant_columns]
```

```python
df = df[df["Erstzulassung"] > 2014]
df.head(10)
```

```python
df["Erstzulassung"].value_counts()
```

```python
df['Kraftstoffart'].value_counts()
```

```python
# Keep only rows where count is the type of Kraftstoff is in one category that has more than 20 elements
value_counts = df['Kraftstoffart'].value_counts()
df = df[df['Kraftstoffart'].isin(value_counts[value_counts >= 20].index)]
```

```python
# harmonize the types of "Kraftstoffart"

kraftstoffart_mapping = {
 'Benzin': 'Benzin', 
 'Diesel': 'Diesel', 
 'Hybr.Benzin/E': 'Hybrid E', 
 'Elektro': 'Elektro', 
 'Hybr.B/E ext.aufl.': 'Hybrid E', 
 'Hybr.Diesel/E': 'Hybrid E', 
 'Benzin/Flüssiggas': 'Hybrid', 
 'Hybr.D/E ext.aufl.': 'Hybrid E', 
 'Erdgas NG': 'Gas', 
 'Benzin/komp.Erdgas': 'Hybrid', 
 }

df['Kraftstoffart'] = df['Kraftstoffart'].replace(kraftstoffart_mapping)
df["Kraftstoffart"].value_counts()
```

```python
# replace this Kraftstoffart that has a value of 0, which in this case is a NAN with the mode.
mode_kraftstoffart = df['Kraftstoffart'].mode().iloc[0]  # Extract the most frequent value
df['Kraftstoffart'] = df['Kraftstoffart'].str.replace('0', mode_kraftstoffart)
df["Kraftstoffart"].value_counts()
```
