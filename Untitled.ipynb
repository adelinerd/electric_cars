```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Sample DataFrame
data = {
    'Category': ['Photovoltaic', 'non-Photovoltaic'],
    'Quantity': [25, 75]
}
df = pd.DataFrame(data)

# User-defined colors
colors = ['#81E552', '#029BD8']

# Plotting
plt.figure(figsize=(6, 6))
plt.pie(df['Quantity'], colors=colors,  startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
plt.show()
```
