```python
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots()

# Plot some data
ax.plot([0, 1, 2, 3], [0, 1, 4, 9])

# Reduce the border thickness by adjusting the linewidth of the axes
#ax.spines['top'].set_linewidth(0.5)   # Top border
ax.spines['right'].set_linewidth(0.8) # Right border
#ax.spines['bottom'].set_linewidth(0.5) # Bottom border
ax.spines['left'].set_linewidth(0.8)  # Left border

# Show the plot
plt.show()
```

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot some data
ax.plot([0, 1, 2, 3], [0, 1, 4, 9])

# Remove the box by hiding the spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Show the plot
plt.show()
```
